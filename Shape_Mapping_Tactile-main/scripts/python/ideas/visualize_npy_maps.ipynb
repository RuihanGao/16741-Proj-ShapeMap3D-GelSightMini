## Crete gelmap for GelSight mini

```python
# Method 1. resize the gelmap2.npy to (240, 320)
import cv2
map3 = cv2.resize(map, (320, 240))
print(f"check map3: {map3.shape}")
plt.imshow(map3)
plt.colorbar()
plt.show()

# save the resized map
map_path = os.path.normpath(os.path.join(os.getcwd(), "../config", "gelmap3.npy"))
np.save(map_path, map3)
```

```python
# Method 2. Since GelSight mini is almost flat, we should expect to get a more uniform map.
# To determine the value, we temporarily set it to be the grayscale value of frame 0
import cv2
# read frame_0.png
frame_0 = cv2.imread("frame_0.png", cv2.IMREAD_GRAYSCALE).astype(np.float32)

# smoothen the image using median blur
frame_0 = cv2.medianBlur(frame_0, 3)
# smoothen the image using gaussian blur
frame_0 = cv2.GaussianBlur(frame_0, (7, 7), 7)

print(f"check frame_0: {frame_0.shape}")
plt.imshow(frame_0)
plt.colorbar()
plt.show()

# save the resized map
map_path = os.path.normpath(os.path.join(os.getcwd(), "../config", "gelmap4.npy"))
np.save(map_path, frame_0)
```
